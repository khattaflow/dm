```python
import pandas as pd
import numpy as np
dataset_1 = pd.read_csv('market_basket.csv', header=None);
dataset_1
```

```python
row,col=dataset_1.shape
print(row,",",col)
```

```python
#preprocessing of dataset_1 to make it  compatible for rule mining.
transactions1=[]
for i in range(0,row):
    temp=[]
    for j in range(0,col):
        # print(dataset_1[i][j])
        if(type(dataset_1.iloc[i][j])==float):
            break
        else:
            temp.append(dataset_1.iloc[i][j])
            
    transactions1.append(temp)
# transactions1
```

```python
#rule mining
itemset1,rules1 = apriori(transactions1,min_support=0.02,min_confidence=0.2)
```

```python
itemset1
```

```python
rules1
```

```python
#processing little bit more to get good result 
for i in transactions1:
    if(len(i)<25):
        transactions1.remove(i)
```

```python
itemset1,rules1=apriori(transactions1,min_support=0.01,min_confidence=0.6)
```
