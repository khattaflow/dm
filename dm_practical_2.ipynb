```python
import pandas as pd
import numpy as np
np.random.seed(0)
data={
    'age': np.random.randint(20, 57, size = 100),
    'income': np.random.randint(20000, 2000000, size = 100)
}
```

```python
df = pd.DataFrame(data)
df
```

```python
# df_standard = (df-df.mean())/df.std()
df_standard.std()
```

```python
df_normalize = (df-df.min())/(df.max()-df.min())
df_normalize
```

```python
df_transformed = df.copy()
df_transformed['income'] = np.log(df_transformed['income'])
df_transformed.head()
```

```python
df_aggregated = df.groupby('age').mean().reset_index()
df_aggregated
```

```python
bins = [0,30,50,np.inf]
labels = ['Young', 'Adult', 'Elderly']
df_discretized = df.copy()
df_discretized['Age Group'] = pd.cut(df_discretized["age"], bins = bins, labels = labels)
df_discretized
```

```python
from sklearn.preprocessing import Binarizer
threshold = 30
binarizer = Binarizer(threshold = threshold)
df_binarized = binarizer.fit_transform(df[['age']])
df_binarized = pd.DataFrame(df_binarized, columns=['Binarized_age'])
df_binarized = pd.concat([df, df_binarized], axis = 1)
print('Binarized dataset: ')
print(df_binarized.head())
```
